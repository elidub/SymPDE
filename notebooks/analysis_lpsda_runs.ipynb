## Losses

```python
def extract_losses(filename):
    normalized_test_loss_values = []

    with open(filename, 'r') as file:
        for line in file:
            if "Normalized test loss mean" in line:
                # Extract the numerical part of the line
                try:
                    value = float(line.split('Normalized test loss mean')[1].split(',')[0].strip())
                    normalized_test_loss_values.append(value)
                except ValueError:
                    print(f"Error extracting value from line: {line}")

    losses = np.array(normalized_test_loss_values)
    return losses
```

```python
def plot_losses(filename, slurm_output_dir = '../jobs/slurm_output/'):
    losses = extract_losses(os.path.join(slurm_output_dir, filename))
    print('# losses = ', len(losses))
    print(losses)
    if len(losses) % 2 != 0: 
        losses = losses[:-1]
    losses = losses.reshape(-1, 2).T

    plt.figure(figsize=(3,4))
    bins = np.linspace(losses.min(), losses.max(), 10)
    _, bins, _ = plt.hist(losses[0], label='Aug', alpha = 0.5, bins = bins)
    plt.hist(losses[1], label='No Aug', alpha = 0.5, bins = bins)
    plt.legend()
    plt.title(filename)
    plt.show()
    return losses
```
