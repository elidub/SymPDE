```python
from itertools import combinations
import numpy as np
from matplotlib import pyplot as plt

def calculate_combinations(elements):
    all_combinations = []
    # Generate all combinations for lengths from 1 to the length of the elements list
    for r in range(1, len(elements) + 1):
        c = combinations(elements, r)
        c = [''.join(combo) for combo in c]
        all_combinations.extend(c)
    return all_combinations

# Elements to combine
elements = [f'g_{i}' for i in range(1, 5)]

# Calculate all combinations
combs = calculate_combinations(elements)
mock_data = [[1-len(combo)/12+0.1 + np.random.normal(scale = 0.1), r'$' + combo + r'$'] for combo in combs]

mock_data  = [[1., r'$\emptyset$']] + mock_data

mock_data
```

```python
with plt.xkcd():
    fig, ax = plt.subplots(figsize = (8, 4))

    for i, (y, label) in enumerate(mock_data):
        x = (len(label)-2)/3 if label != r'$\emptyset$' else 0
        ax.errorbar(x, y, yerr = 1/1*np.abs(np.random.normal(scale = .1)), label = label, fmt = 'o')
        ax.annotate(label, (x, y), xytext = (x + 0.1, y))

    ax.set_xticks(np.arange(0, 5, 1))
    ax.set_xlabel('Number of generators')
    ax.set_ylabel('Test MSE')
    ax.set_title('Mock plot of effect of combinations of generators')
```
